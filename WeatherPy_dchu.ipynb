```python
#import dependencies

import matplotlib.pyplot as plt
import json
import random
import pandas as pd
import requests
```

```python
cities_df = pd.read_csv('Resources/cities.csv')

cities_df.rename(index=str, 
                 columns={'Lng': 'Longitude',
                         'Lat': 'Latitude',
                         'Max Temp': 'Max Temperature',
                         'Cloudiness': 'Cloud Cover'},
                inplace=True)
```

```python
cities_df.to_csv('cities_clean.csv', index=False)
```

```python
#Set sizes

plt.rcParams.update({'figure.figsize': (40,20)})
plt.rcParams.update({'font.size': 40})
```

```python
# World Map, City Coordinate Distribution

cities_df.plot.scatter('Longitude', 
                       'Latitude',
                       title='City Coordinate Distribution',
                       s=400,
                       ylim=[-90,90],
                       color='LightBlue',
                       edgecolor='Black',
                       grid=True) 
```

```python
# Lat v. Temp

cities_df.plot.scatter('Latitude', 
                       'Max Temperature', 
                       title='City Latitude v. Max Temperature', 
                       marker='o', 
                       s=400,
                       xlim=[-90, 90],
                       color='LightBlue', 
                       edgecolor='black', 
                       grid=True)
```

```python
#Lat v. Humidity

cities_df.plot.scatter('Latitude', 
                       'Humidity', 
                       title='City Latitude v. Humidity', 
                       marker='o', 
                       s=400,
                       xlim=[-90, 90],                       
                       color='LightBlue', 
                       edgecolor='black', 
                       grid=True)
```

```python
#Lat v. Clouds

cities_df.plot.scatter('Latitude', 
                       'Cloud Cover', 
                       title='City Latitude v. Cloud Cover', 
                       marker='o', 
                       s=400,
                       xlim=[-90, 90],
                       color='LightBlue', 
                       edgecolor='black', 
                       grid=True)
```

```python
# Lat v. Wind

cities_df.plot.scatter('Latitude', 
                       'Wind Speed', 
                       title='City Latitude v. Wind Speed', 
                       marker='o', 
                       s=400,
                       xlim=[-90, 90],
                       color='LightBlue', 
                       edgecolor='black', 
                       grid=True)
```
